```python
# Import required libraries
import pandas as pd # Data manipulation and analysis
import numpy as np  # Numerical operations
import matplotlib.pyplot as plt # Visualizations
%matplotlib inline
```

```python
# synthetic dataset
X= 6* np.random.rand(100,1) -3
y= 0.5 + X**2 + 1.5*X + 2 + np.random.randn(100,1) # np.random.randn(100,1) is for outliers

# scatter plot
plt.scatter(X,y)
plt.xlabel('X')
plt.ylabel('y')
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)   #random_state ensure same split every time
```

```python
# Implement simple linear regression
from sklearn.linear_model import LinearRegression

lin_reg_model = LinearRegression()
lin_reg_model.fit(X_train, y_train)
```

```python
# R2 score metrics
from sklearn.metrics import r2_score

# calculate score
score = r2_score(y_test,lin_reg_model.predict(X_test) )
print(score)
```

```python
# Lets visualize
plt.plot(X_train, lin_reg_model.predict(X_train), color='red')
plt.scatter(X_train, y_train)
plt.xlabel('X')
plt.ylabel('y')
```

```python
# Lets apply polynomial transformation
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias = True)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.fit_transform(X_test)
```

## PIPELINE CONCEPTS

```python
from sklearn.pipeline import Pipeline
```

```python
def poly_regression(degree):
  X_new = np.linspace(-3, 3, 200).reshape(200, 1)

  poly_features=PolynomialFeatures(degree=degree,include_bias=True)
  lin_reg=LinearRegression()
  poly_regression=Pipeline([
      ("poly_features",poly_features),
      ("lin_reg",lin_reg)
  ])
  poly_regression.fit(X_train,y_train) ## ploynomial and fit of linear reression
  y_pred_new=poly_regression.predict(X_new)
  #plotting prediction line
  plt.plot(X_new, y_pred_new,'r', label="Degree " + str(degree), linewidth=2)
  plt.plot(X_train, y_train, "b.", linewidth=3)
  plt.plot(X_test, y_test, "g.", linewidth=3)
  plt.legend(loc="upper left")
  plt.xlabel("X")
  plt.ylabel("y")
  plt.axis([-4,4, 0, 10])
  plt.show()
```

```python
poly_regression(6)
```

### choosing degree- as more degree- overfitting, less leads to underfitting
- either use cross validation score or RMSE/R2 metrics or both

```python
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np

degrees = [1, 2, 3, 4, 5]
best_score = -np.inf
best_degree = 1

for d in degrees:
    poly = PolynomialFeatures(degree=d)
    X_poly = poly.fit_transform(X_train)
    model = LinearRegression()
    score = cross_val_score(model, X_poly, y_train, cv=5, scoring='r2').mean()
    if score > best_score:
        best_score = score
        best_degree = d

print("Best degree:", best_degree)
```

# Polynomial Regression Explained in Simple Words

Polynomial regression is like linear regression but allows the line to **bend** to fit curves in the data.  
It’s useful when the relationship between your input(s) and output isn’t a straight line.

---

## When to Use Polynomial Regression

You use polynomial regression when:  
- The relationship between inputs and output **isn’t a straight line**.  
- Example: Studying more hours first improves your exam score, but after a point, too much studying makes you tired and the score goes down — a curve, not a line.  
- The data looks like a curve when plotted.  
- Example shapes: U-shaped, upside-down U, or wave-like.  

A straight-line model (simple or multiple linear regression) isn’t enough.  
- Linear regression would miss the bends.  
- Polynomial regression can **“bend the line”** to follow the curve.

---

### ✅ Good Examples
- Predicting plant growth over time (grows fast, slows down, then stops)  
- Predicting car speed vs. fuel efficiency (sweet spot in the middle)  
- Predicting learning rate vs. performance (improves, peaks, then declines)  

### ❌ Not Good to Use If
- The curve is too messy (other models may be needed)  
- Polynomial degree is too high → the curve wiggles too much → overfits  

---

## Think About It

### 1. One Independent Feature (x)
**Degree 1 (Linear Regression):**  
\[
y = \beta_0 + \beta_1 x
\]

**Degree 2 (Quadratic / Polynomial Regression):**  
\[
y = \beta_0 + \beta_1 x + \beta_2 x^2
\]

Where:  
- \(x\) = independent variable  
- \(y\) = dependent variable  
- \(\beta_0, \beta_1, \beta_2\) = coefficients  

---

### 2. Two Independent Features (x₁, x₂)
**Degree 1 (Linear Regression):**  
\[
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2
\]

**Degree 2 (Polynomial Regression):**  
\[
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \beta_3 x_1^2 + \beta_4 x_2^2 + \beta_5 x_1 x_2
\]

- Includes squared terms for each feature (\(x_1^2, x_2^2\))  
- Includes interaction term (\(x_1 x_2\))  

---

### ✅ Summary Table

| Features | Degree | Equation |
|----------|--------|----------|
| 1        | 1      | \(y = \beta_0 + \beta_1 x\) |
| 1        | 2      | \(y = \beta_0 + \beta_1 x + \beta_2 x^2\) |
| 2        | 1      | \(y = \beta_0 + \beta_1 x_1 + \beta_2 x_2\) |
| 2        | 2      | \(y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \beta_3 x_1^2 + \beta_4 x_2^2 + \beta_5 x_1 x_2\) |

---

### ✅ Intuition
- **Low degree → simple → may underfit** if data is curved  
- **Higher degree → more curvy → fits better but risk of overfitting**  
- As you increase the polynomial degree, the curve adapts more to the shape of your data
